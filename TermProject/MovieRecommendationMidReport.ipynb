# 0) 주제 : 나를 위한 영화 추천 프로그램

### 1) 프로젝트 목표 및 내용 

이 프로 그램의 목적은 2018년의 영화들을 추려서, 나의 기준에 맞게 평가하고 점수를 매겨서, 영화를 추천하는 것이다.
나는 영화를 매우 좋아하는데 2018년에는 재수를 하게 되어 영화를 많이 보지 못했다. 
그해에 보지 못하고 지나가니, 그해에 개봉한 영화들 중에 내가 관심이 가는 영화를 찾아보기에는, 막막함을 느껴 미뤄두고 있었다.
이 프로그램을 개발하여, 나에게 2018년중 내가 제일 관심이 갈 만한 영화를 찾고, 그 영화를 여름방학에 시청해보려고 한다.

### 2)주제 선정 이유 또는 프로젝트 필요성

처음에는 지역별 운동시설 현황에 대한 분석과 그에 따른 거주지 선정으로 주제를 선정했는데, 이 주제로 하기에는 범위가 너무 넓고, 지역을 전국구가 아닌 서울로 한정한다 하더라도, 운동에 최적인 거주지를 선정하기 위해서, 각각 아파트 별로 점수를 매기고 비교를 했어야 하는데, 
이는 제출기한안에 해결할 수 없을 것이라는 생각이 들었다. 또, 아파트별로 좌표를 각가 다르게 설정해, 일정 거리내의 운동시설의 유무에 따라서 가중치를 부여하려고 했는데 이부분에 있어서 api를 사용하는 게 아직은 내 역량 밖이라는 생각이 들어서 주제를 나를 위한 영화 추천 프로그램으로 틀게 되었다. 
영화를 좋아하는 나로서는 이런 프로그램을 만들면서, api도 사용해보고, 또 여름방학에 볼 영화를 찾을 수 있다는 점에서 매력적인 프로젝트가 될 것이다. 따라서 흥미를 가지고 임할 수 있을 것 같다. 

### 3) 데이터 획득 

데이터는 네이버 api 영화 카테고리를 사용해서 획득할 예정이다. 그 안에서 ,네티즌 평점, 관람객 평점, 총 관람객수, 러닝타임, 등의 정보를 얻어오고 나의 기준에 맞게 가중치를 두어서 각각에 점수를 부여하고 그 각각의 점수들을 총합해서 점수를 산출하는 방법을 사용하려고 한다.

### 4) 구현 내용 설명

아직 다 구현은 안했기 때문에 계획에 대해서 이야기 하자면, 우선 네이버 api 영화 카테고리에서 2018 년 도 영화에 대한 정보를 얻어 올 것이다
얻어 온 후 1.영화 장르 2.러닝타임  3.관람객 평점 네티즌 평점 4. 총 관람객 수 로 기준을 잡고 비교를 할 것인데, 

1.영화 장르 부분

드라마 8점, 판타지 7점, 서부 3점,
공포 6점, 로맨스 8점, 모헙 8점,
스릴러 8점, 느와르 7점, 컬트 4점,
다큐멘터리 4점, 코미디 8점, 가족 8점, 
미스터리 8점, 전쟁 8점, 애니메이션 5점, 
범죄 8점, 뮤지컬 9점, SF 9점, 
액션 9점, 무협 0점, 에로 0점,
서스펜스 7점, 서사 5점, 블랙코미디 1점, 
실험 4점, 영화카툰 0점, 영화음악 0점, 
영화패러디포스터 0점

이렇게 점수를 부여할 것이고, 

2. 영화 러닝타임 
0분 - 80분      7점
80분 -  120분   10 점
120분 - 160분   9 점
160분 - 180분   7 점
180~분          5 점
으로 부여할 것이고,

3. 관람객 평점, 네티즌 평점, 평론가 평점

(관람객 평점 * 0.5) + (네티즌 평점 * 0.4) + (평론가 평점 * 0.1)
이와 같이 부여 할 것인데, 여기서 네이버 평점의 만점이 10점이기 때문에 다음과 같이 총 점수 합을 구해도 최대 10이 되도록 산출 방식을 적용하였다

4. 총 관람객 수

이부분에 있어서는 아직 기준을 세우는 데 고민중이다. 

1번 기준 + 2번 기준 + 3번 기준 + 4번 기준으로 총합을 구해서 
높은 점수를 얻은 영화를 나에게 추천해주는 방식으로 결과값을 산출할 예정이다.
