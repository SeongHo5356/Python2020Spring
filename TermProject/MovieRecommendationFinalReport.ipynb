# 주제 : 영화 추천받기


### 1) 프로젝트 목표 및 내용

본 프로젝트가 목표로 하는 것은 영화를 크롤링하여 영화 정보를 읽어오고, 기준에 따라 점수를 부여해 영화를 높은 점수부터 나열해 주는
프로그램을 만드는 것이다. 이프로그램을 개발하여 내가 제일 관심이 갈만한 영화를 찾고, 그 영화를 여름방학에 시청하려고 한다.

### 2)  주제 선정 이유 또는 이 프로젝트의 필요성 

나는 영화를 매우 좋아하는데 2018년에는 재수를 하게 되어 영화를 많이 보지 못했다. 
그해에 보지 못하고 지나가니, 그해에 개봉한 영화들 중에 내가 관심이 가는 영화를 찾아보기에는, 막막함을 느껴 미뤄두고 있었다. 
영화를 좋아하는 나로서는 이런 프로그램을 만들면서, 크롤링도 해보고, 또 여름방학에 볼 영화를 찾을 수 있다는 점에서 매력적인 프로젝트가 될 것이다. 따라서 흥미를 가지고 임할 수 있을 것 같다 주제로 영화 추천받기 프로그램 제작을 선정하게 되었다. 중간 보고서에서는 영화 정보를 네이버 api를 이용하여 모은다고 하였다. 네이버 개발자 센터에서 api 승인요청도 하여 부푼 마음으로 api를 다뤄보려고 했지만, 데이터 가공 전에 데이터를 가져오지 조차 못하고 자꾸 코드에 에러가 나서 결국 api를 사용하는 것은 포기하였다. 그 대안으로 찾은 것이 웹 크롤링을 통하여 데이터를 획득하는 것이었다.

### 3) 데이터 획득 

데이터 획득은 위에 2번에서 변경사항에서 이야기 했듯이 웹 크롤링을 통하여 획득하였다. 총 영화에에 관련된 5가지 데이터 0.영화코드 1.영화제목, 2.평점관련, 3.상영시간 관련, 4. 장르 , 5.코멘트를 남긴 사람의 수의 데이터를 수집했다. 우선, 네이버 영화 상세페이지(참고 https://movie.naver.com/movie/bi/mi/basic.nhn?code=182835)를 보게 되면 제일 오른쪽에 영화 코드가 부여되어있다는 것을 보고, 2018년 개봉한 영화들의 코드를 받아오기로 하였다. 위의 참고 페이지같은 개별적인 영화 페이지마다 가져오기에는 너무 시간이 오래 걸릴 것 같아
"https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open=2018" 에서 태그정보에 들어있는 영화 코드를 읽어왔다. <br>
영화 코드의 경우 (ul class="directory_list") 하위의 'li'태그 내부에 (a href='영화상세페이지')같이 존재한다. 따라서 (a href="영화상세페이지") 태그 전체를 str으로 타입캐스트한후, 수기로 영화 코드를 슬라이싱 하는 방법을 택했다.
영화 제목의 경우, (h3 class="h_movie")내부에 텍스트 형태로 존재했기 떄문에, (h3 class="h_movie") 태그를 찾고 거기서 텍스트 정보를 읽어왔다.
영화 평점 관련 정보에 경우 영화 상세페이지에서만 네티즌, 관람객, 평론가, 내 평점 을 구분해서 제공하기 때문에 상세페이지에서 수집하였는다. 평점의 경우, (span class="st_on" style="width:'평점X10'") 같은 형태로 되어 있는 태그에서 가져왔는데, 본래는 빈 별점을 채우는 목적으로 쓰여진 것이라 css의 형태로 들어있어, 영화제목과 같은 형식으로는 얻어올 수 없어다. 따라서, (span class="st_on")을 가진 태그를 읽어와서 문자열로 타입캐스트한 후에 수기로 width값을 슬라이싱 해와서 얻었다. 여기서 문제가 있었던 것은 비주류 영화의 경우 평론가의 평점이 존재하지 않는 경우가 있었다. 이 경우 내 평점을 평론가 평점으로 대체하기로 했다.
상영시간 관련해서는, (p class="info_spec")태그 내부에 (span)태그를 통해 크롤링하였다. 장르또한 (p class="info_spec")내부의 (span)태그 안에 존재하여 같은 방식으로 대신 장르는 한영화당 여러개가 존재할 수 있기때문에, 리스트 형태로 받아왔다.
코멘트를 남긴 사람의 수는 (span class="user_count")내부에 (em)태그안에 텍스트 형태롤 존재했기 떄문에, 제목과 같은 방식으로 텍스트를 긁어왔다. 
    
### 4) 구현 내용 설명 
  
전체젝인 구현 방식은 네이버영화 목록에서 영화코드를 읽어온 후 각각의 페이지에 들어가서 정보를 읽어온 후, 새로운 객체로 각각의 영화를 등록한다. 등록을 마친 후, 점수를 산출하여 순위를 결정하는 방식이다. 이를 구현하기 위해서 우선, 클래스를 통해 recommendation 객체를 설정했주었다. 점수에 영향을 주는 것은 4가지였는데, 1. 영화 장르, 2.영화 런타임, 3.평점 , 4.코멘트를 남긴 사람의 수이다.


영화 장르의 경우, 메소드 중 genreScore()로 산출하였고,기준은 <br>
드라마 8점, 판타지 7점, 서부 3점,<br> 
공포 6점, 로맨스 8점, 모헙 8점,<br>
스릴러 8점, 느와르 7점, 컬트 4점,<br>
다큐멘터리 4점, 코미디 8점, 가족 8점, <br>
미스터리 8점, 전쟁 8점, 애니메이션 5점, <br>
범죄 8점, 뮤지컬 9점, SF 9점, <br>
액션 9점, 무협 0점, 에로 0점,<br>
서스펜스 7점, 서사 5점, 블랙코미디 1점, <br>
실험 4점, 영화카툰 0점, 영화음악 0점, <br>
영화패러디포스터 0점<br>
로 영화 점수를 부여했다. 한 영화가 장르를 다양하게 가질 수 있었기 때문에, 영화가 가진 장르당 점수를 모두 더하고 평균 내는 형식으로 하였다. 장르에서의 만점은 10점이다.

영화 런타임의 경우, 메소드 중 runTimeScore()로 산출했고, 기준은 <br>

으로 부여하였다. 영화 런타임에서의 만점 또한 10점다.  <br>
    
영화 평점의 경우,starScore()로 산출했고, 기준은 아래와 같다<br>
기본적으로 네티즌 평점은 0.2, 평론가 평점 또한 0.2, 관람객 평점 0.6의 가중치를 주어, 합을 구했다.
따라서 영화 평점 부분 또한 만점은 10점이다
    
    
영화 코멘트를 남긴 사람의 수의 경우, 메소드 중 participantsScore()로 산출했다. 원래 기준은 총 관람객 수 였지만 바뀐 이유로는, 총관람객수를 네이버 영화에서 제공하는 것을 찾을 수 없었기 때문이다 같은 맥락에서 코멘트를 남긴 사람의 수는 사람들의 관심도를 측정할 수 있다고 생각했기 때문에, 대체할 기준으로 뽑았다.<br>

영화 추천 순위를 결정하는 총점의 경우 위 네가지 기준으로 산출한 점수를 모두 더하고 난 후 * 10/4 를 통하여 총점의 최대를 100점으로 맞춰주었다. 크롤링해온 평점의 경우, 비주류 영화는 평론가 평점이 없어, 평론가 평점을 상세 페이지 중 내 평점으로 바꾸어 적용하였다. 여기서 문제는 평점을 읽어올떄, 평론가 평점이 존재하는 경우, 관람객 평점, 평론가 평점, 네티즌 평점을 순서대로 긁어왔었는데, 평론가 평점이 없다보니, 이 순서가 관람객 평점, 네티즌 평점, 내평점 순으로 읽어졌다. 따라서, 점수를 산출하는 데 문제가 없도록, 평론가평점(내평점)에 부여하는 가중치와, 네티즌 평점에 부여하는 가중치를 동일하게 0.2로 두어 해결하였다. 동작방법으로는 밑에 MovieForMe(year, page)에서 내가 찾고 싶은 년도 영화와 , 몇페이지까지의 영화를 크롤링해와 비교할 것인지 결정할 수 있다. (https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open=+"year"+&page=+"page")에서 괄호 url 속에 "page","year"속에 입력 파라메터가 들어가서 크롤링하여 영활를 추천해준다. 
또한 조금 수정을 거치면, 굳이 년도에서 몇페이지까지 읽어와 추천해주는 프로그램이 아닌, 내가 원하는 영화들을 집어넣어 내가 고민이 되는 영화들이 있을 떄 점수를 내어 좀더 나의 흥미와 일치하는 영화를 찾을 수 있을 것 같다는 점에서 확장성이 있는 것 같다.

### 6) 결론 

우선 나는 내프로그램은 아직 단점이 많다. 거의 모든 정보를 태그를 찾고 그 태그를 문자열로 타입캐스트 한 후, 슬라이싱 했기 떄문에, 만약에 앞에 다른 정보가 추가 또는 삭제 되어 내가 찾으려는 태그가 한 인덱스씩 밀리게 된다면 수기로 찾아줘야 한다. 이런 점이 영화코드를 획들 할 떄도 단점으로작용했다. 목록 중 영화의 장르가 명시되지 않은 영화들의 경우, 태그가 밀리게 되어, 영화코드를 제대로 받아오지 못하여 이런 영화들은 수기로 영화 코드를 찾아 영화목록의 후보 중에서 제외시켜줘야 한다. 이렇게 제외시켜줘야 하는 영화의 목록이 나와있는 게  MovieForMe() 내부의 remove_movies리스트이다. (15페이지까지에서만 해도 나에게 추천할 만한 영화후보가 충분히 많다고 생각되어, 현재 15페이지까지 remove_movies 리스트를 등록해놓았다.) 또한, 영화코드를 긁어오고 긁어온 영화 코드와 지울 영화코드를 대조하여 영화코드 목록을 업데이트 한후 영화목록에 있는 각각의 고유영화코드에 따라 영화상세페이지에 접속해 정보를 읽어와야 하기 뗴문에, 실행 시간이 너무 많이 걸린다는 단점이 있다. 그러나, 내가 목표로 했던, 나를 위한 영화 추천이라는 목표에는 도달했다. 수기로 정보를 찾고 비교할 경우 상상도 못할 만큼의시간이 걸릴텐데 20분 정도가 걸리는 프로그램의 러닝타임은 참아줄만 한다는 생각이다. 그러나 최적화에 대한 개선은 필요하다.

```python
import urllib.request
from bs4 import BeautifulSoup


class Recommendation:
    
    #객체 생성자
    def __init__(self,movieScore,title, genres, runTime, viewScore, netScore, score, participants):
        self.title = title
        self.movieScore = movieScore
        self.genres = genres
        self.runTime = runTime
        self.viewScore = viewScore
        self.netScore = netScore
        self.score = score
        self.participants = participants
    
    #객체정보를 보여주는 method
    def show(self):
        print("제목 : ",self.title)
        print("추천점수 : ",self.movieScore)
        print("장르 : " ,self.genres )
        print("상영시간 : " , self.runTime)
        print("관람객평점 : ",float(self.viewScore)/10)
        print("관심을 표한 사람 수 : ",self.participants)
    
    # 장르마다 점수를 부여하는 method
    def genreScore(self):
        genreScore_list=[]
        for genre in self.genres:
            if (genre=="드라마"):
                genreScore_list.append(8)
            elif(genre =="판타지"):
                genreScore_list.append(7)
            elif(genre =="서부"):
                genreScore_list.append(3)
            elif(genre =="공포"):
                genreScore_list.append(6)
            elif(genre =="로맨스"):
                genreScore_list.append(8)
            elif(genre =="모험"):
                genreScore_list.append(8)
            elif(genre =="스릴러"):
                genreScore_list.append(8)
            elif(genre =="컬트"):
                genreScore_list.append(4)
            elif(genre =="다큐"):
                genreScore_list.append(4)
            elif(genre =="코미디"):
                genreScore_list.append(8)
            elif(genre =="가족"):
                genreScore_list.append(8)
            elif(genre =="미스터리"):
                genreScore_list.append(8)
            elif(genre =="전쟁"):
                genreScore_list.append(8)
            elif(genre =="애니메이션"):
                genreScore_list.append(5)
            elif(genre =="범죄"):
                genreScore_list.append(8)
            elif(genre =="뮤지컬"):
                genreScore_list.append(9)
            elif(genre =="SF"):
                genreScore_list.append(9)
            elif(genre =="액션"):
                genreScore_list.append(9)
            elif(genre =="무협"):
                genreScore_list.append(0)
            elif(genre =="에로"):
                genreScore_list.append(0)
            elif(genre =="서스펜스"):
                genreScore_list.append(7)
            elif(genre =="서사"):
                genreScore_list.append(5)
            elif(genre =="블랙코미디"):
                genreScore_list.append(1)
            elif(genre =="실험"):
                genreScore_list.append(4)
            else:
                genreScore_list.append(0)
        
        #리스트에 있는 총점 계산
        sum =0
        for genreScore in genreScore_list:
            sum += genreScore

        return sum/len(self.genres)
    
    #상영시간에 따라서 영화점수 부여 
    def runTimeScore(self):
        
        if(0 <= float(self.runTime) and float(self.runTime) < 80):
            runTIme_Score = 7
        elif(80 <= float(self.runTime) and float(self.runTime) < 120):
            runTIme_Score = 10
        elif(120 <= float(self.runTime) and float(self.runTime) < 160):
            runTIme_Score = 9
        elif(160 <= float(self.runTime) and float(self.runTime) <180):
            runTIme_Score = 7
        else:
            runTIme_Score = 5
            
        return runTIme_Score
    
    #영화의 평점에 따른 점수 계산
    def starScore(self):
        
        #가끔 평론가의 평점이 관람객과 너무 다를 떄가 있기때문에 가중치 0.02
        #받아온 정보가 별점 9.06 이면 90.6 으로 받아왔기 떄문에 ex) 0.2X  0.02 o
        #인기가 없는 영화중에는 평론가의 평점이 없는 영화도 있으 그럴떄는 내 평점(무조건 0점)으로 평론가 평점을 대체함
        movie_star_score = float(self.viewScore) * (0.06) + float(self.netScore) *(0.02) + float(self.score)* (0.02)
        
        return movie_star_score
    
    #영화에 대한 관심도(평가한 네티즌 + 관람객수)에 따른 점수부여
    def participantsScore(self):
        scroe = 0
        if (float(self.participants) < 5000):
            score = 8.0
        elif(5000 <= float(self.participants) and float(self.participants) < 10000):
            score = 8.2
        elif(10000 <= float(self.participants) and float(self.participants) < 15000):
            score = 8.5
        elif(15000 <= float(self.participants) and float(self.participants) < 20000):
            score = 8.8
        elif(20000 <= float(self.participants) and float(self.participants) < 25000):
            score = 9
        elif(25000 <= float(self.participants) and float(self.participants) < 30000):
            score = 9.5
        else:
            score = 10
        
        return score
    
    #영화의 추천점수(총점을 계산)
    def calMovieScore(self):
        
        genre_score = self.genreScore()
        runtime_score = self.runTimeScore()
        star_score = self.starScore()
        numviewer_score = self.participantsScore()
        
        moviescore = genre_score + runtime_score + star_score + numviewer_score
        
        #40점 만점의 점수를 100점 만점으로 변환(기준당 10점)
        self.movieScore = moviescore*10/4
    
    #객체의 str 타입의 점수를 float 타입으로 casting
    def scoreTypeCast(self):
        self.movieScore = float(self.movieScore)
    
    #객체의 총점을 받아오는 함수
    def getMovieScore(self):
        return self.movieScore
    
        
```

```python
def crawlMovies(year,page):
    total_list=[]
    for i in range(1, page+1):
        url = "https://movie.naver.com/movie/sdb/browsing/bmovie.nhn?open="+str(year)+"&page="+str(i)
        soup = BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")
        mlist = soup.find("ul",class_="directory_list")
    
        new_title_list=[]
        title_list=[]   
        needList2 = mlist.select("li")
        for i in range(0,len(needList2)):
            a = str(needList2[i])
            b = a[5:12]
            if (b =="<a href"):
                title_list.append(a)

    
        for i in range(0, len(title_list)):
            a = title_list[i]
            b = a[42:48]
            if (b[5]=='"'):
                continue
            new_title_list.append(b)
        
        total_list+=new_title_list
    return total_list
```

```python
#웹에 접근해서 정보를 읽어오는 함수
def crawl(url):
    
    soup = BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")
    title = soup.find('h3', class_='h_movie').find('a').text

    # html의 태그중 <span 태그중 st_on을 클래스로 가지는 태그를 뽑아온다.
    ds = soup.select('span.st_on')
    
    # 최초로 나온 세개는 관람객 평점, 네티즌 평점, 평론가 평점 정보를 가지고 있다 -> 태그 안쪽에 width:로 정보가 존재하기 떄문에,
    # 태그를 문자열로 바꾸어 평점 정보가 위치하는 부분만을 잘라서 가져온다.
    #여기서 뽑아온 rank 값은 평점에 10이 곱해져 있다는것을 유의하고 나중에 점수를 산출할 떄 보정하도록 한다. (평점 : 9.06 --> 90.6)
    for d in range(0,3):
        a = str(ds[d])
        rank = a[33:37]
        if (d == 0):
            if(rank == '0%">' or rank =='0.0%'):
                rank = "0"
            RviewScore = rank
        elif (d == 1):
            if(rank == '0%">' or rank =='0.0%'):
                rank = "0"
            Rscore = rank
        elif(d == 2):
            if(rank == '0%">' or rank =='0.0%'):
                rank = "0"
            RnetScore = rank
            
            
    #영화의 상영시간을 크롤링한다
    ms = soup.find("p", class_ ="info_spec")
    runTime = ms.select("span")[2].text[:-2]
    
    
    #장르를 크롤링해온 후 태그,여백정보는 자르고 장르정보만 받아온다,
    genres = ms.select("span")[0].text.strip().split(sep = ',')
    for i in range(0, len(genres)):
        genres[i]=genres[i].replace("\r",'')
        genres[i]=genres[i].replace('\t','')
        genres[i]=genres[i].replace('\n','')
        genres[i]=genres[i].replace(' ','')
    
    #평가에 참여한 네티즌 ,관람자의 수(관심도)
    participants = soup.find("span", class_="user_count").find("em").text
    new_participants = participants.replace(",", "")  
    
    
    
    
    
    return Recommendation(0, title, genres, runTime, RviewScore, Rscore, RnetScore, new_participants)

crawl("https://movie.naver.com/movie/bi/mi/basic.nhn?code=36944")
```

```python
def findMovie(movies_list, remove_movies):
    for i in movies_list:
        for k in remove_movies:
            if (i ==k):
                movies_list.remove(k)
    return movies_list

# 순위를 정할 영화 리스트
def MovieForMe(year, page):
    movies_list = crawlMovies(year,page)
    print(movies_list)
    # 영화장르가 옴니버스 영화로 영화 상세페이지에서 장르르 제공하지 않거나 영화 장르가 불분명한 영화들 목록-->기준 미달
    remove_movies=['171837','175274', '173968', '173419', '174012', '173420', '174032', '174030', '174013', '162981', '181563', '174039', '174443', '174038', '172518', '172506', '172520', '181562', '173470', '166416', '173451', '174031','173470']
   

    movies_list= findMovie(movies_list, remove_movies)

    # 영화데이터를 Recommendation 객체 형태로 담을 리스트 
    recommendations_list = []

    #영화 정보 크롤링 후 객체 형태의 영화 정보를 recommendations_list에 append
    for movie in movies_list:
        recommendation = crawl('https://movie.naver.com/movie/bi/mi/basic.nhn?code=' + str(movie))
        recommendations_list.append(recommendation)

    #영화의 순위비교를 위해 총점 계산 후 str-->float으로 typecast    
    for r in recommendations_list:
        r.calMovieScore()
        r.scoreTypeCast()

    #영화의 점수가 높은 영화부터 차례대로 출력
    for r in sorted(recommendations_list, key=lambda recommendation : recommendation.movieScore, reverse = True):
        r.show()
        print()
    
```
